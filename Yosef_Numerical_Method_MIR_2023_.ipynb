### 1.3 Dichotomy


Also known as binary search, is a popular algorithmic technique used to solve optimization problems. It involves dividing the search space in half at each iteration and eliminating one half of the search space based on a binary decision until a satisfactory solution is found. This technique is commonly used to find the minimum or maximum value of a function within a given interval.

To use dichotomy, we first define an interval in which the optimal solution is expected to lie. We then divide the interval in half and evaluate the function at the midpoint. If the function value at the midpoint is greater than the value at one of the endpoints, we continue the search in the half of the interval that includes the endpoint with the lower function value. Otherwise, we continue the search in the other half of the interval.

This process is repeated until the interval becomes sufficiently small or the desired accuracy is achieved. The final solution is typically taken as the midpoint of the final interval. Dichotomy is an efficient and reliable optimization technique, particularly for one-dimensional optimization problems. It is guaranteed to converge to the optimal solution in a finite number of steps and can be easily implemented in software using a loop or recursion.

However, dichotomy can be less efficient than other optimization techniques for high-dimensional problems, as the number of function evaluations required can grow exponentially with the dimensionality of the search space.


Sure, here are the steps to find the zeros of a function using the dichotomy method:

 

1.   Define the interval [a, b] that contains the root of the function f(x). This interval should be chosen such that f(a) and f(b) have opposite signs (i.e., f(a) * f(b) < 0). This ensures that there is at least one root in the interval.
2.   Calculate the midpoint of the interval, c = (a + b) / 2.
3.   Evaluate the function at the midpoint, f(c).
4.   Determine which half of the interval contains the root by checking the sign of f(c). If f(c) has the same sign as f(a), then the root must be in the interval [c, b]. Otherwise, the root must be in the interval [a, c].
5.   Repeat steps 2-4 until the interval [a, b] is sufficiently small or the desired accuracy is achieved.
6.   The final solution is the midpoint of the final interval, (a + b) / 2.


```python
# 3. Find the zero of these functions by Dichotomy.

def dichotomy(f, a, b, tol):
    # Iterate until the interval width is less than the tolerance
    while (b - a) > tol:
        # Calculate the midpoint of the interval
        c = (a + b) / 2
        # If the function value at the midpoint is zero, we have found the root
        if f(c) == 0:
            return c
        # If the signs of the function values at the endpoints and midpoint are different, the root is in the left half of the interval
        elif f(a) * f(c) < 0:
            b = c  # Update the right endpoint of the interval
        # Otherwise, the root is in the right half of the interval
        else:
            a = c  # Update the left endpoint of the interval
    # The midpoint of the final interval is the estimated root
    return (a + b) / 2
```

### 1.4. Find the zero of these functions by Newton Raphson.

The Newton-Raphson method, also known as the Newton method, is a popular iterative optimization algorithm used to find the roots of a function or the minimum/maximum of a function. It is a powerful optimization method that is commonly used in engineering, science, and finance.

The Newton-Raphson method works by starting with an initial guess for the root or optimal solution and then iteratively refining that guess using the derivative of the function. The method assumes that the function is differentiable and that its derivative is continuous.

At each iteration, the method computes the slope of the tangent line at the current guess and then finds the point where the tangent line intersects the x-axis. This point is taken as the new guess for the root or optimal solution. The process is then repeated until a satisfactory solution is found.

The algorithm can be mathematically represented as:

$x_{n+1} = x_n - f(x_n)/f'(x_n)$

where x_n is the current guess for the root or optimal solution, $f(x_n)$ is the value of the function at x_n, and $f'(x_n)$ is the derivative of the function at $x_n$.

The Newton-Raphson method is a powerful optimization method that can converge very quickly if the initial guess is close to the true root or optimal solution. However, it can also converge very slowly or even fail to converge if the function is highly non-linear or has multiple roots/optima. Additionally, the method may not work well if the function is not differentiable or if the derivative is difficult to compute.

Overall, the Newton-Raphson method is a useful optimization technique that is widely used in various fields. It is a powerful tool for finding roots or extrema of a function, but it requires careful consideration of its limitations and proper implementation to ensure its success.


Sure, here are the step-by-step instructions for finding the zeros of a function using the Newton-Raphson method:

1. Choose an initial guess x0 for the root of the function.

2. Calculate the value of the function f(x0) and its derivative f'(x0) at the initial guess x0.

3. Calculate the next guess x1 using the formula:

      x1 = x0 - f(x0)/f'(x0)

4. Check if the absolute difference between x1 and x0 is smaller than a predefined tolerance. If it is, stop the iteration and return x1 as the root. If it is not, set x0 = x1 and go back to step 2.

5. If the iteration has not converged after a maximum number of iterations, stop the algorithm and return an error message.

6. If the function has multiple roots or a singular point, the algorithm may fail to converge to the desired root. In this case, try different initial guesses or consider using a different method.

Note that step 2 requires the calculation of the derivative of the function. This can be done analytically if a closed-form expression for the function is available, or numerically using finite difference approximations. In the latter case, it is important to choose a sufficiently small step size to ensure accurate results.

Also, note that the convergence of the Newton-Raphson method depends on the initial guess and the properties of the function. In some cases, the method may converge slowly or even fail to converge. Therefore, it is important to choose a reasonable initial guess and to verify the convergence of the method for the specific function being considered.

```python
def newton_raphson(f, f_prime, x0, tol=1e-6, max_iter=100):
    """
    Implementation of the Newton-Raphson method for finding the root of a function.

    Parameters:
        f (function): The function for which to find the root.
        f_prime (function): The derivative of the function.
        x0 (float): The initial guess for the root.
        tol (float): The tolerance for the absolute difference between successive iterates.
        max_iter (int): The maximum number of iterations.

    Returns:
        float: The approximate root of the function.
    """
    # Iterate until the maximum number of iterations is reached or the tolerance is met
    for i in range(max_iter):
        fx = f(x0)
        fx_prime = f_prime(x0)

        # Calculate the next guess using the Newton-Raphson formula
        x1 = x0 - fx / fx_prime

        # Check if the tolerance is met
        if abs(x1 - x0) < tol:
            return x1

        # Update the guess for the next iteration
        x0 = x1

    # Return an error message if the maximum number of iterations is reached
    raise ValueError("Newton-Raphson method did not converge within the specified maximum number of iterations.")
```
